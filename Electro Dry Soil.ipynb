```python
%pylab inline
pylab.rcParams['figure.figsize'] = (8.0,6.0)
```

```python
mu=4*pi*1e-7
ep=3*8.8541878176e-12
s=1e-5 

#alfa general
def ag(f):
    return (2*pi*f)*sqrt(1/2.*mu*ep*(sqrt(1+(s/(2*pi*f*ep))**2)-1))

#alfa conductor
def ac(f):
    return sqrt(pi*mu*s*f)

#alfa dielectrico
ad=(s/2.)*sqrt(mu/ep)

#beta general
def bg(f):
    return (2*pi*f)*sqrt(1/2.*mu*ep*(sqrt(1+(s/(2*pi*f*ep))**2)+1))

#beta conductor
def bc(f):
    return sqrt(pi*mu*s*f)

#beta dielectrico
def bd(f):
    return 2.*pi*f*sqrt(mu*ep)


f=linspace(1,3e5,10000) #en Hz

```

# Alphas

```python
pd,pg,pc = plot(f,ad*f/f,f,ag(f),f,ac(f))

legend((r'$\alpha \, good \, dielectric$',r'$\alpha \,general$',r'$\alpha \, good \, conductor$'),loc=0,fancybox=True)
title(r'$\alpha\,of\,dry\,soil\,vs\,frequency$')
xlabel(r'$Frequency\,[Hz]$')
ylabel(r'$\alpha \, [\frac{Np}{m}]$')
xlim(0,250000)
ylim(0,0.0025)
```

# Betas

```python
qd,qg,qc = plot(f,bd(f),f,bg(f),f,bc(f))

legend((r'$\beta \, good \, dielectric$',r'$\beta \,general$',r'$\beta \, \, good \, conductor$'),loc='upper left',fancybox=True)
title(r'$\beta\,of\,dry\,soil\,vs\,frequency$')
xlabel(r'$Frequency\,[Hz]$')
ylabel(r'$\beta \, [\frac{rad}{m}]$')
xlim(0,200000)
ylim(0,0.007)
```

# Deltas

```python
ppd,ppg,ppc = plot(f,1./ad*f/f,f,1./ag(f),f,1./ac(f))

legend((r'$\delta \, good \, dielectric$',r'$\delta \,general$',r'$\delta \,  good  \, conductor$'),loc=(0.5,0.6),fancybox=True)
title(r'$Penetration\,depth\,of\,dry\,soil\,vs\,frequency$',verticalalignment='bottom')
xlabel(r'$Frequency\,[Hz]$')
ylabel(r'$\delta \, [\frac{m}{Np}]$')
xlim(0,50000)
ylim(0,6000)
```

# Velocities

```python
vd,vg,vc = plot(f,2*pi*f/bd(f),f,2*pi*f/bg(f),f,2*pi*f/bc(f))

legend((r'$v \, good \, dielectric$',r'$v \,general$',r'$v \,  good  \, conductor$'),loc='best',fancybox=True)
title(r'$Velocity\,of\,the\, wave\,vs\,frequency$',verticalalignment='bottom')
xlabel(r'$Frequency\,[Hz]$')
ylabel(r'$Velocity \, [\frac{m}{s}]$')
xlim(0,200000)
ylim(0,4e8)
```

# Quotients

```python
plot(f,(s*(ad*f/f)**-1)**-1,label=r'$\delta\,good\,dielectric$')
plot(f,(s*(ag(f))**-1)**-1,label=r'$\delta\,general$')
plot(f,(s*(ac(f))**-1)**-1,label=r'$\delta\,good\,conductor$')
xlabel(r'$Frequency\,[Hz]$')
ylabel(r'$Quotient\,\frac{1}{\sigma \delta} \, [ \frac{Np}{S\,m} ] $')
title(r'$Proporcionality\,quotients\,vs\,frequency$')
xlim(0,250000)
ylim(0,200)
legend(loc='best',fancybox=True)
```
